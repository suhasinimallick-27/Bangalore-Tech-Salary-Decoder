```python
import pandas as pd
df = pd.read_csv("bangalore_tech_salaries.csv")
print(df.head())
print(df.tail())
print(df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.duplicated().sum())





```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ","_")

print(df.columns)
```

```python
df = df.drop_duplicates()
print("Duplicate rows:", df.duplicated().sum())
```

```python
# REMOVE EXTRA SPACES FROM ALL COLUMS
for col in df.select_dtypes(include="object"):
  df[col]= df[col].str.strip()
  print(df.isnull().sum())
```

```python

# Standardize Company Type

df["company_TYPE"] = df["company_TYPE"].replace({
    "UNICORN": "Unicorn",
    "unicorn": "Unicorn",
    "MNC": "MNC",
    "mnc": "MNC",
    "MID-SIZE": "Mid-size",
    "Mid Size": "Mid-size",
    "EARLY-STAGE": "Early-stage",
    "Early Stage": "Early-stage"
})

print(df["company_TYPE"].value_counts())
```

```python
# Remove commas from Current CTC
df["current_ctc"] = df["current_ctc"].str.replace(",", "", regex=False)

# Remove LPA
df["current_ctc"] = df["current_ctc"].str.replace("LPA", "", regex=False)

# Remove spaces
df["current_ctc"] = df["current_ctc"].str.strip()

# Convert to number
df["current_ctc"] = pd.to_numeric(df["current_ctc"], errors="coerce")
```

TASK 3

```python
#Q1
# AVERAGE , MEDIAN, MIN, MAX, SALARY BY ROLES

role_salary = df.groupby("role")["current_ctc"].agg(
    Mean = "mean",
    Median = "median",
    minimum = "min",
    maximum = "max"
)

role_salary = role_salary.sort_values(by="Median", ascending = False)
print(role_salary)
```

```python
print(df.columns)
```

```python
#Q2                                     # HERE I USED AI

# Re-standardize column names to handle potential state inconsistencies
df.columns = df.columns.str.lower()
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ","_")

# Re-clean and convert 'current_ctc' to numeric, as previous steps might not have persisted.
df["current_ctc"] = df["current_ctc"].astype(str) # Ensure it's string before operations
df["current_ctc"] = df["current_ctc"].str.replace(",", "", regex=False)
df["current_ctc"] = df["current_ctc"].str.replace("LPA", "", regex=False)
df["current_ctc"] = df["current_ctc"].str.replace("₹", "", regex=False) # Remove currency symbol
df["current_ctc"] = df["current_ctc"].str.strip()
df["current_ctc"] = pd.to_numeric(df["current_ctc"], errors="coerce")

# Filter only SDE Backend employees
backend = df[df["role"] == "SDE Backend"].copy()

# Create experience bands
backend.loc[:, "exp_band"] = pd.cut(
    backend["years_exp"],
    bins=[0, 1, 3, 5, 100],
    labels=["0-1", "2-3", "4-5", "6+"]
)

# Median salary by experience band
exp_salary = backend.groupby("exp_band")["current_ctc"].median()
print(exp_salary)
```

```python
#Q3
# Compare salaries of employees with and without specific skills.
# This helps identify which skills increase salary.
sde = df[
    df["role_clean"].isin(
        ["SDE Backend",
         "SDE Frontend",
         "SDE Full-Stack"]
    )
]

skills = ["AWS","ML","System Design","Kubernetes"]

for skill in skills:

    with_skill = sde[
        sde["skills"]
        .str.contains(skill,
                      case=False,
                      na=False)
    ]["current_ctc_lpa"].median()

    without_skill = sde[
        ~sde["skills"]
        .str.contains(skill,
                      case=False,
                      na=False)# Check whether a particular skill exists in the skills column.
# str.contains helps search text values.
    ]["current_ctc_lpa"].median()

    print(skill)

    print("With:",with_skill)

    print("Without:",without_skill)
```

```python
#Q4
# Median salary by company type
company_salary = df.groupby("company_type")["current_ctc"].median()
print(company_salary)
```

```python
#Q5
# Median salary by role
df["role_median"] = df.groupby("role")["current_ctc"].transform("median")

# Salary Gap
df["salary_gap"] = df["role_median"] - df["current_ctc"]

# Top 10 underpaid employees
underpaid = df.sort_values("salary_gap", ascending=False)
print(underpaid[["employee_id", "role", "current_ctc", "salary_gap"]].head(10))
```
